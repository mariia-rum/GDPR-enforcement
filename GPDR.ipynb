```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
GDPR = pd.read_excel("Book1.xlsx")
```

```python
GDPR.describe()
```

```python
GDPR.info()
```

```python
GDPR['Country'].value_counts()
```

```python
_ = GDPR['Country'].value_counts().plot(kind = 'bar', color = 'g', alpha=0.5 , grid=True)
plt.ylabel("Number of fines")
plt.xlabel("Country")
plt.title("Histogram of the GDPR fines in the EU countries (May 2018-June 2020)")
plt.show()
```

```python
_ = GDPR['Offender'].value_counts().plot(kind = 'bar', color = 'y', alpha=0.5 , grid=True, label = None)
plt.ylabel("Numbers of fines")
plt.title ("Histogram representing parties who have breached the GDPR (May 2018-June 2020)")
plt.legend(loc='upper right')
plt.show()
```
